```python
import matplotlib.pyplot as plt
from glob import glob
from tqdm import tqdm
import cv2
import os
import seaborn
```

```python
path = r'C:\Users\user\AIHub_seg\datasets\person_crop'

img_list = glob(os.path.join(path, '*.jpg'))
# for tmp in img_list:
#     print(tmp)
print(len(img_list))
```

```python
h_list = []
w_list = []
```

```python
for img in img_list:
    src = cv2.imread(img)
    h,w,c = src.shape
    h, w, c = src.shape
    h_list.append(h)
    w_list.append(w)
    
print('person class 총 개수 : ', len(img_list))
```

```python
print('\nh_min : ', min(h_list))
print('h_max : ', max(h_list))
print('\nw_min : ', min(w_list))
print('w_max : ', max(w_list))
```

```python
plt.hist(h_list, bins=25, edgecolor='black')
plt.title('Height')
plt.xlabel('Size')
plt.ylabel('Count')
plt.axis([0, 1200, 0, 60000])
plt.show()
```

```python
plt.hist(w_list, bins=25,edgecolor='black')
plt.title('Width')
plt.xlabel('Size')
plt.ylabel('Count')
plt.axis([0, 1200, 0, 60000])
```
