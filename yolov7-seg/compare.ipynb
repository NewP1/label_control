```python
def createDirectory(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print("Error: Failed to create the directory.")
```

```python
original_path = r'C:\Users\user\yolov8\datasets\labels\val'
o_label_list = glob(os.path.join(original_path,'*.txt'))

compare_path = r'C:\Users\user\yolov8\runs\detect\resize_0.25\labels'
c_label_list = glob(os.path.join(compare_path,'*.txt'))

# print(o_label_list[0])
# print(o_label_list[0][len(original_path)+1:])
# print(c_label_list[0])
# print(c_label_list[0][len(compare_path)+1:])
```

```python
def make_label(file_name):
    label_result = []
    with open(file_name,'r') as f:
        text = f.readlines()
    for line in text:
        if(int(line.split()[0]) in des_classes):
            label_result.append(int(line.split()[0]))  # classes[int(line.split()[0])]
    return sorted(label_result)
```

```python
des_num = 1
compare_num = None
total_count = len(o_label_list)
c_count = 0 # 정검지 카운트
f_count = 0 # 오검지 카운트

dc_txt = [] # 공존 카운트
diff_txt = []   # 개수 다른거 카운트
f_list_1 = []   # 있는데 없다
f_list_2 = []   # 없는데 있다

for o_label, c_label in tqdm(zip(o_label_list, c_label_list)):
    original = Counter(make_label(o_label))
    result = Counter(make_label(c_label))

    
    if des_num in original :
        if des_num in result:
            if original[des_num] > result[des_num]:
                diff_txt.append(o_label[len(original_path)+1:])
                f_count += 1
            else:
                c_count += 1
        else:
            f_list_1.append(o_label[len(original_path)+1:])
            f_count += 1
    
    else:
        if des_num in result:

            f_list_2.append(o_label[len(original_path)+1:])
            f_count += 1
        else:
            c_count += 1
    
    if compare_num is not None:
        if des_num in original and compare_num in original:
            if des_num in result and compare_num in result:
                dc_txt.append(o_label[len(original_path)+1:])

fail_list = diff_txt + f_list_1 + f_list_2              
print("\ndes_num 개수 적음 : ", len(diff_txt))
for tmp in diff_txt:
    print(tmp)

print("\ndes, com 같이 있음 (고민해 볼 부분): ", len(dc_txt))
for tmp in dc_txt:
    print(tmp)
    
print("\n있는데 없다 : ", len(f_list_1))
for tmp in f_list_1:
    print(tmp)
    
print("\n없는데 있다 : ", len(f_list_2))
for tmp in f_list_2:
    print(tmp)


print('\n총 개수 : ', len(o_label_list))
print("정검지 카운트 : ", c_count)  
print("오검지 카운트 : ", f_count)
print("accuracy : ", round(c_count/total_count * 100, 2))
  
```
