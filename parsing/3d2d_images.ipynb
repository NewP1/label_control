```python
import os
from glob import glob
import cv2
import json
```

```python
def createDirectory(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print("Error: Failed to create the directory.")
```

```python
path = r'C:\Users\user\Desktop\의류 가상착용 3D이미지\test'
p = glob(os.path.join(path, '*.json'))
print(p)

des_path = os.path.join(path, 'labels_txt')
createDirectory(des_path)
```

```python
classes = {'shirt' : 0, 'vest' : 3, 'outer' : 2, 'dress' : 8, 'shorts' : 5, 'long pants' : 4, 'casual pants' : 4, 'jeans' : 4, 'mini skirt' : 6, 'long skirt' : 6, 'casual skirt' : 6
           , 'jean skirt' : 6, 'baseball cap' : 10, 'beret' : 10, 'sun-cap' : 10, 'beanie' : 10, 'sneakers' : 12, 'combat boots' : 12, 'rain boots' : 12, 'sandals' : 12, 'slippers' : 12, 'running shoes' : 12
           , 'formal shoes' : 12, 'neckerchief' : 14, 'scarf' : 14, 'hair band' : 11}

# shirt, outer, sun-cap, formal shoes 확인 필요, pajamas, stuff 확인 필요
```

```python
for file in p:  
    result = []
    with open(file) as f:
        text = json.load(f)
    t_name = file[len(path)+1:-5] + '.txt'

    # print(t_name)

    line = text['shapes'][0]
    
    
    shape_type = line['shape_type']

    if shape_type == 'polygon':
        label = line['label']

        if label in classes:
            l_id = classes[label]
            result.append(str(l_id))
            points = line['points']
            h, w = int(text['imageHeight']), int(text['imageWidth'])
            for point in points:
                result.append(str(round(point[0]/w, 6)))
                result.append(str(round(point[1]/h, 6)))
            print(result)
            with open(os.path.join(des_path, t_name), 'w') as f:
                f.write(' '.join(result)+'\n')

    
```
